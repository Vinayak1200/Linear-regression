Import necessary libraries

```python
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Load dataset using pydataset module

```python
pima = data('Pima.tr')    #load and view the dataset
print(pima)         

pima.plot(kind = 'scatter', x = 'skin', y = 'bmi')
plt.show()
```

split the dataset into testing and training data

```python
x_train, x_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
```

plot the training and testing dataset

```python
plt.scatter(x_train, y_train, label = 'training data', color = 'r', alpha=0.7)
plt.scatter(x_test, y_test, label='testing data', color = 'g')
plt.legend()
plt.title("Test train split")
plt.show()
```

Creating a linear model

```python
LR = LinearRegression()
LR.fit(x_train.values.reshape(-1, 1), y_train.values)
```

Use model to predict on test data

```python
prediction = LR.predict(x_test.values.reshape(-1, 1))

#plot prediction line and compare it with test data
plt.plot(x_test, prediction, label='Linear Regression', color='b')
plt.scatter(x_test, y_test, label='actual test data', color='g')
plt.legend()
plt.show()
```

Scoring the model

```python
LR.score(x_test.values.reshape(-1, 1), y_test.values)
```
